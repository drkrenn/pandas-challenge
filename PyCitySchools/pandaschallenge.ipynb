```python
import csv
import pandas as pd
from pathlib import Path
```

```python
schools_path = Path("Resources", 'schools_complete.csv')
schools_df= pd.read_csv(schools_path)
schools_df
```

```python
school_list = schools_df["school_name"].unique()
school_count = schools_df["school_name"].count()
school_count
```

```python
total_students= schools_df["size"].sum()
total_students
```

```python
total_budget = schools_df["budget"].sum()
total_budget
```

```python
students_path = Path("Resources", 'students_complete.csv')
students_df= pd.read_csv(students_path)
students_df
```

```python
mathaverage = students_df["math_score"].mean()
mathaverage=mathaverage.round(2)
mathaverage
```

```python
readaverage = students_df["reading_score"].mean()
readaverage=readaverage.round(2)
readaverage
```

```python
passmath_df= students_df.loc[students_df["math_score"]>60]
passmath_df
```

```python
math_pass_students= passmath_df["math_score"].count()
math_percent = math_pass_students/total_students*100
math_percent=math_percent.round(2)
math_percent
```

```python
passread_df= students_df.loc[students_df["reading_score"]>60]
passread_df
```

```python
read_pass_students= passread_df["reading_score"].count()
read_percent = read_pass_students/total_students*100
read_percent=read_percent.round(2)
read_percent
```

```python
overall_percent = math_percent
overall_percent
```

```python
district_summary = {'School List': [school_count], 
                      'Total Students': [total_students],
                   'Total Budget': [total_budget],
                   'Average Math Score': [mathaverage], 
                   'Average Reading Score': [readaverage],
                   'Percent Passing Math': [math_percent],
                   'Percent Passing Reading': [read_percent],
                   'Percent Passing Both': [overall_percent]}
district_summary_df = pd.DataFrame(data=district_summary)
district_summary_df
```

```python
sumschools_df = schools_df[["school_name", "type", "size", "budget"]]
sumschools_df.head()
```

```python
sumschools_df["per_student_budget"] = sumschools_df["budget"]/sumschools_df["size"]
sumschools_df
```

```python
school_groups= students_df.groupby(["school_name"])
school_avg_df= school_groups.mean("math_score", "reading_score")

school_avg_df
```

```python
passmath_school=passmath_df.groupby(["school_name"])
passmath_school_df = passmath_school.count()
passmath_school_df =passmath_school_df["math_score"]
passmath_school_df
```

```python


school_merged_df = pd.merge(sumschools_df, school_avg_df, on=["school_name"])
school_merged_df
```

```python
school_merged_df = pd.merge(school_merged_df, passmath_school_df, on=["school_name"])
school_merged_df
```

```python
school_merged_df = school_merged_df.rename(columns={"math_score_y":"mathpass", "math_score_x":"math_score"})
school_merged_df
```

```python
school_merged_df["percent_passed_math"]=school_merged_df["mathpass"]/school_merged_df["size"]*100
school_merged_df["percent_passed_reading"]=100
#note: previously identified that everyone passed reading. Hence, reading percentage is 100% for all schools, and percent that passed both would be equal to percent that passed math. 
school_merged_df["percent_passed_both"]=school_merged_df["percent_passed_math"]
school_merged_df
```

```python
school_summary_final_df = school_merged_df[['school_name', 'type', 'size', 'budget', 'per_student_budget', 'reading_score', 'math_score', 'percent_passed_math', 'percent_passed_reading', 'percent_passed_both']]
school_summary_final_df = school_summary_final_df.rename(columns={"school_name":"School Name", "type":"School Type",  "size":"Total Students",  "budget":"School Budget",  "per_student_budget":"Budget- Per Student",  "reading_score":"Average Reading Score",  "math_score":"Average Math Score",  "percent_passed_math":"Percent Passed Math",  "percent_passed_reading":"Percent Passed Reading",  "percent_passed_both":"Overall Passing"})
school_summary_final_df
```

```python
top_schools =school_summary_final_df.sort_values(by='Overall Passing', ascending=False).head(5)
top_schools
```

```python
bottom_schools=school_summary_final_df.sort_values(by='Overall Passing', ascending=True).head(5)
bottom_schools
```

```python
ninth_df= students_df.loc[students_df["grade"]==str("9th")]
tenth_df= students_df.loc[students_df["grade"]==str("10th")]
eleventh_df= students_df.loc[students_df["grade"]==str("11th")]
twelfth_df= students_df.loc[students_df["grade"]==str("12th")]
```

```python
ninereadaverage=ninth_df["reading_score"].mean()
ninereadaverage=ninereadaverage.round(2)
tenreadaverage=tenth_df["reading_score"].mean()
tenreadaverage=tenreadaverage.round(2)
elevenreadaverage=eleventh_df["reading_score"].mean()
elevenreadaverage=elevenreadaverage.round(2)
twelvereadaverage=twelfth_df["reading_score"].mean()
twelvereadaverage=twelvereadaverage.round(2)
```

```python
ninemathaverage=ninth_df["math_score"].mean()
ninemathaverage=ninemathaverage.round(2)
tenmathaverage=tenth_df["math_score"].mean()
tenmathaverage=tenmathaverage.round(2)
elevenmathaverage=eleventh_df["math_score"].mean()
elevenmathaverage=elevenmathaverage.round(2)
twelvemathaverage=twelfth_df["math_score"].mean()
twelvemathaverage=twelvemathaverage.round(2)
```

```python
mathbygrade = {'9th Grade Math Average': [ninemathaverage], 
                 '10th Grade Math Average': [tenmathaverage],
                 '11th Grade Math Average': [elevenmathaverage],
                '12th Grade Math Average': [twelvemathaverage]}
mathbygrade_df = pd.DataFrame(data=mathbygrade)
mathbygrade_df
```

```python
readbygrade = {'9th Grade Read Average': [ninereadaverage], 
                 '10th Grade Read Average': [tenreadaverage],
                 '11th Grade Read Average': [elevenreadaverage],
                '12th Grade Read Average': [twelvereadaverage]}
readbygrade_df = pd.DataFrame(data=readbygrade)
readbygrade_df
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_spending_df = school_summary_final_df
school_spending_df["Spending Ranges Per Student"] = pd.cut((school_summary_final_df['Budget- Per Student']), spending_bins, labels=labels)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges Per Student"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges Per Student"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges Per Student"])["Percent Passed Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges Per Student"])["Percent Passed Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges Per Student"])["Overall Passing"].mean()
```

```python
spending_summary = pd.merge(spending_math_scores, spending_reading_scores, on=["Spending Ranges Per Student"])
spending_summary
spending_summary = pd.merge(spending_summary, spending_passing_math, on=["Spending Ranges Per Student"])
spending_summary
spending_summary = pd.merge(spending_summary, spending_passing_reading, on=["Spending Ranges Per Student"])
spending_summary
spending_summary = pd.merge(spending_summary, overall_passing_spending, on=["Spending Ranges Per Student"])
spending_summary
```

```python

size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
size_school_df = school_summary_final_df
size_school_df["School Size"] = pd.cut((school_summary_final_df['Total Students']), size_bins, labels=labels)
```

```python
size_math_scores = size_school_df.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = size_school_df.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = size_school_df.groupby(["School Size"])["Percent Passed Math"].mean()
size_passing_reading = size_school_df.groupby(["School Size"])["Percent Passed Reading"].mean()
overall_passing_size = size_school_df.groupby(["School Size"])["Overall Passing"].mean()
```

```python
size_summary = pd.merge(size_math_scores, size_reading_scores, on=["School Size"])
size_summary = pd.merge(size_summary, size_passing_math, on=["School Size"])
size_summary = pd.merge(size_summary, size_passing_reading, on=["School Size"])
size_summary = pd.merge(size_summary, overall_passing_size, on=["School Size"])
size_summary
```

```python
type_math_scores = school_summary_final_df.groupby(["School Type"])["Average Math Score"].mean()
type_reading_scores = school_summary_final_df.groupby(["School Type"])["Average Reading Score"].mean()
type_passing_math = school_summary_final_df.groupby(["School Type"])["Percent Passed Math"].mean()
type_passing_reading = school_summary_final_df.groupby(["School Type"])["Percent Passed Reading"].mean()
overall_passing_type = school_summary_final_df.groupby(["School Type"])["Overall Passing"].mean()
```
